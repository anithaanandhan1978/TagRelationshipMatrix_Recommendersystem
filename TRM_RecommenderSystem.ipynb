```python
import csv
import xml.etree.ElementTree as ET

def xml_to_csv(xml_file, csv_file):  
    tree = ET.parse(xml_file)
    root = tree.getroot()

    with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
        csv_writer = csv.writer(csvfile)
       
        csv_writer.writerow(['Id','PostId','VotetypeId','CreationDate'])

        for row in root.findall('row'):
            row_data = [
                row.get('Id'),
                row.get('PostId'),
                row.get('VotetypeId'),
                row.get('CreationDate')                        
            ]
            csv_writer.writerow(row_data)

if __name__ == "__main__":
    xml_file_path = 'G:\\Anitha\\stackexchange\\Votes.xml'  
    csv_file_path = 'G:\\Anitha\\stackexchange\\Votes.csv'      

    xml_to_csv(xml_file_path, csv_file_path)
```
